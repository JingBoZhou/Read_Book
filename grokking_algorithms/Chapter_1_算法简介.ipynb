## 1.2 二分查找

二分查找是一种算法，其输入是一个有序的元素列表，如果要查找的元素包含在列表中，二分查找返回其位置；否则返回null



下面的示例说明了二分查找的工作原理。我随便想一个1～100的数字。



你的目标是以最少的次数猜到这个数字。你每次猜测后，我会说小了、大了或对了。
